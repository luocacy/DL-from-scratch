# 类

#### 实现一个man类

```python
#coding:utf-8
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello " + self.name + "!") 
        
    def goodbye(self):
        print("Goodbye " + self.name + "!")
```

```python
m = Man("Tiko")
m.hello()
m.goodbye()
```

#### 2. 广播
通过扩展，使不同形状的数组间也可以进行运算

```python
B = np.array([[1, 2], [3, 4]])
C = np.array([10, 20])
B * C
```

#### 3. 访问元素

```python
for row in B:
    print(row)
```

```python
B = B.flatten()    #转为一维数组
print(B)
B[np.array([0, 3])]    #获取索引为0，3的元素
```

```python
B > 2
```

```python
B[B > 1]    #找出值大于1的项
```

# Matplotlib

#### 1. 绘制sin函数

```python
import matplotlib.pyplot as plt
```

```python
#生成数据
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
```

```python
#绘制图形
plt.plot(x, y1, label="sin", lineStyle="--")
plt.legend()    #显示代码中label中的内容
plt.show()
```

#### 2. 绘制sin和cos

```python
#生成数据
y2 = np.cos(x)
```

```python
#绘图
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", lineStyle="--")
plt.title("sin & cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### 3. 显示图像
